```python
# Q 1
import numpy as np
# Create a matrix
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
# i. Sum of all elements
total_sum = np.sum(gfg)
print("Sum of all elements:", total_sum)
# ii. Sum of all elements row-wise
row_wise_sum = np.sum(gfg, axis=1)
print("Row-wise sum:\n", row_wise_sum)
# iii. Sum of all elements column-wise
column_wise_sum = np.sum(gfg, axis=0)
print("Column-wise sum:\n", column_wise_sum)
```

```python
# Q 2
import numpy as np
# Given arrays
array1 = np.array([10, 52, 62, 16, 16, 54, 453])
array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
# (i) Sorted array
sorted_array = np.sort(array1)
print("Sorted array:", sorted_array)
# (ii) Indices of sorted array
sorted_indices = np.argsort(array1)
print("Indices of sorted array:", sorted_indices)
# (iii) 4 smallest elements
smallest_elements = np.sort(array1)[:4]
print("4 smallest elements:", smallest_elements)
# (iv) 5 largest elements
largest_elements = np.sort(array1)[-5:]
print("5 largest elements:", largest_elements)
# (b) Filtering integer and float elements
integer_elements = array2[array2 == array2.astype(int)]
float_elements = array2[array2 != array2.astype(int)]
# (i) Integer elements only
print("Integer elements:", integer_elements)
# (ii) Float elements only
print("Float elements:", float_elements)
```

```python
# Q 3
import numpy as np
# Step a: Generate unique sales dataset based on initials
first_initial = ord('R')  # Replace 'R' with your first initial
last_initial = ord('B')   # Replace 'B' with your last initial
X = first_initial + last_initial
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales Data:", sales)
# Step b: Compute personalized tax rate and apply it
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
print("Sales after applying tax:", taxed_sales)
# Step c: Adjust sales based on discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Sales after discount:", discounted_sales)
# Step d: Expand sales data for multiple weeks
weekly_sales = np.vstack([sales, sales * 1.02, sales * 1.04])
print("Weekly sales data:\n", weekly_sales)
```

```python
# Q 4
import numpy as np
import matplotlib.pyplot as plt
# Generate x values
x = np.linspace(-10, 10, 100)
# Compute y values
y_square = x ** 2
y_sin = np.sin(x)
y_exp = np.exp(x)
y_log = np.log(np.abs(x) + 1)
# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, y_square, label='y = x^2', linestyle='dashed')
plt.plot(x, y_sin, label='y = sin(x)', linestyle='dotted')
plt.plot(x, y_exp, label='y = e^x', linestyle='solid')
plt.plot(x, y_log, label='y = log(|x| + 1)', linestyle='dashdot')
# Add title, labels, and grid
plt.title("Function Plots using NumPy and Matplotlib")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)
# Show the plot
plt.show()
```
