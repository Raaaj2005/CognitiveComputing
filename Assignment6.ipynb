```python
# Q 1
import numpy as np
import matplotlib.pyplot as plt
# Ask user for a value M
M = float(input("Enter a value for M: "))
# Generate x values
x = np.linspace(-10, 10, 100)
# Compute y values
y_square = M * (x ** 2)
y_sin = M * np.sin(x)
# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, y_square, label=f'y = {M} * x^2', linestyle='dashed', color='b')
plt.plot(x, y_sin, label=f'y = {M} * sin(x)', linestyle='solid', color='r')
# Add title, labels, legend, and grid
plt.title("Function Plots with User Input M")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)
# Show the plot
plt.show()
```

```python
# Q 2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Create a dataset
data = {
    'Subjects': ['Math', 'Science', 'English', 'History', 'Art'],
    'Scores': [85, 92, 78, 88, 95]
}
# Convert to DataFrame
df = pd.DataFrame(data)
# Plot using Seaborn
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')
# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=12, fontweight='bold')
# Add title, labels, and grid
plt.title("Subject Scores", fontsize=14)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show plot
plt.show()
```

```python
# Q 3
import numpy as np
import matplotlib.pyplot as plt
# Set seed using roll number
np.random.seed(102317090)
# Generate dataset
data = np.random.randn(50)
# Create a figure with 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
# Line plot showing cumulative sum
axes[0, 0].plot(np.cumsum(data), color='b', linestyle='-', marker='o')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)
# Scatter plot with random noise
axes[0, 1].scatter(range(50), data, color='r', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)
# Hide empty subplots
axes[1, 0].axis("off")
axes[1, 1].axis("off")
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Q 4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Replace 'path_to_your_file' with the actual path where the CSV file is located
file_path = "C:\Users\HP\Downloads\company_sales_data.csv"
df = pd.read_csv(file_path)
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    sns.lineplot(x='month_number', y=product, data=df, label=product, marker='o')
plt.title('Sales Data of Different Products Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(title='Products')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='month_number', y='total_profit', data=df, palette='viridis')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(axis='y')
plt.show()

for product in products:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='month_number', y=product, data=df, palette='muted')
    plt.title(f'{product.capitalize()} Sales per Month')
    plt.xlabel('Month Number')
    plt.ylabel(f'{product.capitalize()} Sales')
    plt.grid(axis='y')
    plt.show()
```
