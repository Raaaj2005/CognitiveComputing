```python
# Q 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Part I: Randomized Sales Data Generation
np.random.seed(102317090)  # Roll number as seed
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales_data = np.random.randint(1000, 5000, size=(12, 4))
# Convert to DataFrame
df = pd.DataFrame(sales_data, columns=categories, index=months)
# Part II: Data Manipulation and Analysis
print(df.head())  # Display first 5 rows
print(df.describe())  # Summary statistics
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100  # Percentage change
if 102317090 % 2 == 0:
    df['Electronics'] *= 0.9  # 10% discount for even roll number
else:
    df['Clothing'] *= 0.85  # 15% discount for odd roll number
# Part III: Visualizations
plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Q 2
import numpy as np
# Given array
array = np.array([[1, -2, 3], [-4, 5, -6]])
# i. Element-wise absolute value
abs_array = np.abs(array)
print("Element-wise absolute value:\n", abs_array)
# Flattened array
flattened = array.flatten()
# ii. Percentiles
percentiles_flat = np.percentile(flattened, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
print("\n25th, 50th, and 75th percentile of flattened array:", percentiles_flat)
print("25th, 50th, and 75th percentile for each column:\n", percentiles_col)
print("25th, 50th, and 75th percentile for each row:\n", percentiles_row)
# iii. Mean, Median, and Standard Deviation
mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)
mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("\nMean, Median, and Std of flattened array:", mean_flat, median_flat, std_flat)
print("Mean, Median, and Std for each column:\n", mean_col, median_col, std_col)
print("Mean, Median, and Std for each row:\n", mean_row, median_row, std_row)
```

```python
# Q 3
import numpy as np
# Given array
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
# Floor values
floor_values = np.floor(a)
# Ceiling values
ceil_values = np.ceil(a)
# Truncated values
trunc_values = np.trunc(a)
# Rounded values (to nearest integer)
rounded_values = np.round(a)
print("Original Array:", a)
print("Floor Values:", floor_values)
print("Ceiling Values:", ceil_values)
print("Truncated Values:", trunc_values)
print("Rounded Values:", rounded_values)
```

```python
# Q 4
# Function to swap two elements in a list using a temporary variable
def swap_elements(lst, index1, index2):
    temp = lst[index1]  # Store the first element in a temporary variable
    lst[index1] = lst[index2]  # Assign the second element to the first index
    lst[index2] = temp  # Assign the temporary variable to the second index
    return lst
# Example list
my_list = [10, 20, 30, 40, 50]
# Indices of elements to swap
index1, index2 = 1, 3  # Swapping elements at index 1 and 3
# Swapping elements
swapped_list = swap_elements(my_list, index1, index2)
# Output the swapped list
print("List after swapping:", swapped_list)
```

```python
# Q 5
# Function to swap two elements in a set
def swap_elements_in_set(s, elem1, elem2):
    # Convert set to list
    lst = list(s)
    # Get indices of elements to swap
    if elem1 in lst and elem2 in lst:
        index1, index2 = lst.index(elem1), lst.index(elem2)
        # Swap elements
        lst[index1], lst[index2] = lst[index2], lst[index1]
        # Convert back to set
        return set(lst)
    else:
        print("One or both elements not found in the set.")
        return s
# Example set
my_set = {10, 20, 30, 40, 50}
# Elements to swap
elem1, elem2 = 20, 40
# Swapping elements
swapped_set = swap_elements_in_set(my_set, elem1, elem2)
# Output the swapped set
print("Set after swapping:", swapped_set)
```
